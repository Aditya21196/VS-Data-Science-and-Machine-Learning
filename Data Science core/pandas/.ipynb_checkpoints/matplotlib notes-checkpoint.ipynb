```python
fig = plt.figure()
ax = fig.add_subplot(111)

#setting limits
ax.set_xlim([0.5, 4.5])
ax.set_ylim([-2, 8])

# setting axis labels
ax.set_title('Title')
ax.set_ylabel('Y-Axis description')
ax.set_xlabel('X-Axis description')

# adding data
plt.scatter([2,3.5,1],[1,-1,4],c='yellow')
plt.scatter([1,1.5,1],[-1,1,2],c='b')
plt.show()
```

```python
arr = 10 * np.random.randn(20)
arr
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# setting axis labels
ax.set_title('Title')
ax.set_ylabel('Y-Axis description')
ax.set_xlabel('X-Axis description')

ax.set_xlim([-1, 20])

plt.plot(arr,c = 'yellow')
plt.show()
```

```python
# linspace: (start,stop,number of values)..... produces "number of values" between start and stop inclusive
np.linspace(1,3,4)
```

```python
# creating the sine curve
import math # inbuilt module
math.sin(math.pi/2) #checking up sine function
```

```python
math.pi
```

```python
xt = np.linspace(1,4 * math.pi,100)
xt
```

```python
# list comprehension
arr = [1,2,3]

# list comprehension generates a new array from an original array by performing the transformation intended
# syntax for list comprehension: ["expression" for "iterating object" in "given list"]
new_arr = [x**2 for x in arr]
new_arr
```

```python
# obtaining sine values
yt = [math.sin(x) for x in xt]
```

```python
# plotting the sine values
plt.plot(yt)
plt.show()
```
